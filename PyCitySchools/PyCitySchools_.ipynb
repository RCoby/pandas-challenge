```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_file = "schools_complete.csv"
student_data_file = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_file)
student_data = pd.read_csv(student_data_file)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="outer", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
#Review requirements of Local Government Area (LGA) Summary report to establish data requirements
#Clean the merged dataset/df to enable metric retreval for the Local Government Area (LGA) Summary
#----

#Determine columns required for analysis *Drop - student_name, gender
SchoolsData_Cln = school_data_complete[["Student ID",
                                        "year",
                                        "school_name",
                                        "reading_score",
                                        "maths_score",
                                        "School ID",
                                        "type",
                                        "size",
                                        "budget"]]

#Format Headers 
SchoolsData_Cln=SchoolsData_Cln.rename(columns={"year":"Year",
                                                "school_name":"School Name",
                                                "reading_score":"Reading Score",
                                                "maths_score":"Maths Score",
                                                "type":"School Type",
                                                "size":"Total Students",
                                                "budget":"Total School Budget"})

##Review df
SchoolsData_Cln.head()
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Create COMPLETE core Local Government Area (LGA) df
#------------------------------------------
# METRICS for Local Government Area (LGA) df 

LGA_SchoolData=pd.DataFrame(SchoolsData_Cln)

#Set metrics for column values

# Total School Count
Schools=LGA_SchoolData["School ID"].value_counts()
SchoolCount=Schools.count()
LGA_SchoolData["Total Schools"]=SchoolCount

# Total Student Count
StudentCount=LGA_SchoolData["Student ID"].count()


# Per Student Budget
StudentBudget=LGA_SchoolData.apply(lambda x:x["Total School Budget"] / x["Total Students"],axis=1)
LGA_SchoolData["Per Student Budget"]=StudentBudget

# Total Budget
BudgetSum=StudentBudget.sum()

# Maths Score
MathsScore=LGA_SchoolData["Maths Score"]

# Reading Score
ReadScore=LGA_SchoolData["Reading Score"]

# Av.Math Score
MathScore_Av=LGA_SchoolData["Maths Score"].mean()
LGA_SchoolData["Average Maths Score"]=round(MathScore_Av,2)

# Av.Read Score
ReadScore_Av=LGA_SchoolData["Reading Score"].mean()
LGA_SchoolData["Average Reading Score"]=round(ReadScore_Av,2)

# % Students Passing Math
PassMath=(LGA_SchoolData["Maths Score"]>=50).value_counts()
PassMath_Pcnt=(PassMath[True]/StudentCount.sum())*100
LGA_SchoolData["% Passing Maths"]=round(PassMath_Pcnt,2)

# % Students Passing Read
PassRead=(LGA_SchoolData["Reading Score"]>=50).value_counts()
PassRead_Pcnt=(PassRead[True]/StudentCount.sum())*100
LGA_SchoolData["% Passing Reading"]=round(PassRead_Pcnt,2)


# % Students Passing Math AND Read
PassMandR=((LGA_SchoolData["Maths Score"] >= 50) & (LGA_SchoolData["Reading Score"] >= 50)).value_counts()
PassMandR_Pcnt=(PassMandR[True]/StudentCount.sum())*100
LGA_SchoolData["% Overall Passing"]=round(PassMandR_Pcnt,2)


LGA_SchoolData.head()
```

```python
#Local Government Area (LGA) Summary Report

#Columns for report
area_summary =LGA_SchoolData.head(1)[["Total Schools","Total Students","Total School Budget","Average Maths Score","Average Reading Score","% Passing Maths","% Passing Reading","% Overall Passing"]]

#format output
area_summary ["Total Students"]=f"{StudentCount:,}"
area_summary ["Total School Budget"]=f"${BudgetSum:,.2f}"

area_summary 
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
#create new df for required data using existing LGA_SchoolData df
per_school_summary =LGA_SchoolData.groupby(["School Name","School Type"]).agg({
    "Total Students":"count",           #Total Students
    "Total School Budget":"mean",       #Total School Budget
    "Per Student Budget":"mean",        #Per Student Budget
    "Maths Score":"mean",               #Average Maths Score 
    "Reading Score":"mean",})           #Average Reading Score 

#rename Maths/Reading columns as the output is the average 
per_school_summary =per_school_summary .rename(columns={"Maths Score":"Average Maths Score",
                              "Reading Score":"Average Reading Score"})


#calc % Students Passing Math and round output to 2 decimals
PassMath_P=LGA_SchoolData.groupby(["School Name","School Type"])["Maths Score"].apply(lambda x: (x>=50).mean()*100)
per_school_summary ["% Passing Maths"]=PassMath_P

#calc % Students Passing Read and round output to 2 decimals
PassRead_P=LGA_SchoolData.groupby(["School Name","School Type"])["Reading Score"].apply(lambda x: (x>=50).mean()*100)
per_school_summary ["% Passing Reading"]=PassRead_P

#calc % Students Passing Math AND Read and round output to 2 decimals
PassMandR_P=LGA_SchoolData[(LGA_SchoolData["Maths Score"]>=50) & (LGA_SchoolData["Reading Score"]>=50)].groupby(["School Name","School Type"])["Maths Score"].count()/LGA_SchoolData.groupby(["School Name","School Type"])["Maths Score"].count()*100
per_school_summary ["% Overall Passing"]=PassMandR_P


#format the output to 2 decimals 
per_school_summary=per_school_summary.round(2)

per_school_summary 
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#utilising SchoolSummay df display top 5 performing schools by % overall passing
#sort values
top_schools =per_school_summary.sort_values(by=["% Overall Passing"],ascending=False)

#format output to better present data
top_schools ["Total School Budget"]=top_schools ["Total School Budget"].map("${:,.2f}".format)
top_schools ["Per Student Budget"]=top_schools ["Per Student Budget"].map("${:,.2f}".format)
top_schools ["% Passing Maths"]=top_schools ["% Passing Maths"].map("{:.2f}%".format)
top_schools ["% Passing Reading"]=top_schools ["% Passing Reading"].map("{:.2f}%".format)
top_schools ["% Overall Passing"]=top_schools ["% Overall Passing"].map("{:.2f}%".format)
top_schools ["Total Students"]=top_schools ["Total Students"].map("{:,}".format)

top_schools .head(5)
```

* Sort and display the five worst-performing schools by % overall passing.

```python
#utilising SchoolSummay df display bottom 5 performing schools by % overall passing
#sort values
bottom_schools=per_school_summary.sort_values(by=["% Overall Passing"])

#format output to better present data
bottom_schools["Total School Budget"]=bottom_schools["Total School Budget"].map("${:,.2f}".format)
bottom_schools["Per Student Budget"]=bottom_schools["Per Student Budget"].map("${:,.2f}".format)
bottom_schools["% Passing Maths"]=bottom_schools["% Passing Maths"].map("{:.2f}%".format)
bottom_schools["% Passing Reading"]=bottom_schools["% Passing Reading"].map("{:.2f}%".format)
bottom_schools["Total Students"]=bottom_schools["Total Students"].map("{:,}".format)

bottom_schools.head(5)
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#use existing df and select required columns to create a table listing the av. maths/reading score for each year level
ScoresByYr=LGA_SchoolData[["School Name","Year","Maths Score","Reading Score"]]
```

```python
#MATHS table listing the av. score for each year level

#create year 9 Maths Score df  
y_9=ScoresByYr["Year"]==9
year9_df=ScoresByYr[y_9].groupby(["School Name"])[["Maths Score"]].mean()

#create year 10 Maths Score df  
y_10=ScoresByYr["Year"]==10
year10_df=ScoresByYr[y_10].groupby(["School Name"])[["Maths Score"]].mean()

#create year 11 Maths Score df  
y_11=ScoresByYr["Year"]==11
year11_df=ScoresByYr[y_11].groupby(["School Name"])[["Maths Score"]].mean()

#create year 12 Maths Score df  
y_12=ScoresByYr["Year"]==12
year12_df=ScoresByYr[y_12].groupby(["School Name"])[["Maths Score"]].mean()

#combine scores from each year into a single df
maths_scores_by_year =year9_df.rename(columns={"Maths Score":"Year 9"})
maths_scores_by_year ["Year 10"]=year10_df[["Maths Score"]]
maths_scores_by_year ["Year 11"]=year11_df[["Maths Score"]]
maths_scores_by_year ["Year 12"]=year12_df[["Maths Score"]]


#round output to 2 decimals 
maths_scores_by_year=maths_scores_by_year.round(2)
maths_scores_by_year 
```
